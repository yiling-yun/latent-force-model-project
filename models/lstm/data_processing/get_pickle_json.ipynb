## add force

```python
mat = loadmat("./../../force/rst/all_1133/estpart_forcemodel.v3_all.mat")
mat
```

```python
# Adjust this key depending on your variable name in MATLAB
estpara = mat['estpara'] 
```

```python
# === 2. Convert MATLAB cell array to Python list ===
# Each cell contains an ndarray of shape (n_frames, 11)
force_para = []
for i in range(estpara.size):
    para = estpara[0, i]  # access cell (since MATLAB stores as 2D)
    if para.size == 0:
        continue  # skip empty
    para_np = np.array(para)

    # Example: split columns (adjust column meaning as needed)
    row = {
        'epsilon_selfA': para_np[:, 0].tolist(),
        'sigma_selfA': para_np[:, 1].tolist(),
        'bcoef_selfA': para_np[:, 2].tolist(),
        'epsilon_inter': para_np[:, 3].tolist(),
        'sigma_inter': para_np[:, 4].tolist(),
        'bcoef_inter': para_np[:, 5].tolist(),
        'dev_selfAinter': para_np[:, 6].tolist(),
        'epsilon_selfB': para_np[:, 7].tolist(),
        'sigma_selfB': para_np[:, 8].tolist(),
        'bcoef_selfB': para_np[:, 9].tolist(),
        'dev_selfB': para_np[:, 10].tolist(),
    }
    force_para.append(row)
```

```python
df_force = pd.DataFrame(force_para)
df_force
```

```python
df_force = pd.DataFrame(force_para)
df_traj_force = pd.concat([df_traj, df_force], axis=1)
```
